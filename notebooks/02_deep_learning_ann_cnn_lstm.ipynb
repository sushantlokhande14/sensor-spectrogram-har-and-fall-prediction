```python
import pandas as pd
import numpy as np

def load_mobifall_file(path):
    with open(path, "r", errors="ignore") as f:
        lines = f.readlines()
    start = next(i for i, l in enumerate(lines) if l.strip() == "@DATA") + 1
    rows = []
    for line in lines[start:]:
        line = line.strip().rstrip(",")
        if not line: continue
        parts = [p.strip() for p in line.split(",")]
        if len(parts) != 4: continue
        try:
            rows.append([int(parts[0]), float(parts[1]),
                         float(parts[2]), float(parts[3])])
        except ValueError:
            continue
    return pd.DataFrame(rows, columns=["t_ns", "x", "y", "z"])

sample = f"{ROOT}/sub1/FALLS/FOL/FOL_gyro_1_2.txt"
df = load_mobifall_file(sample)
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 3 rows:")
print(df.tail(3))

dt_ms = np.diff(df["t_ns"].values) / 1e6
print(f"\nΔt stats (ms): mean={dt_ms.mean():.2f}, "
      f"median={np.median(dt_ms):.2f}, "
      f"min={dt_ms.min():.2f}, max={dt_ms.max():.2f}")
print(f"→ effective sample rate: ~{1000/dt_ms.mean():.0f} Hz")
print(f"→ duration: {(df['t_ns'].iloc[-1] - df['t_ns'].iloc[0]) / 1e9:.2f} s")
```

```python
def load_trial(root, subject, category, activity, trial_no, target_hz=100):
    """Load acc + gyro for one trial, resample to target_hz, return (T, 6) array."""
    acc_path  = f"{root}/{subject}/{category}/{activity}/{activity}_acc_{subject[3:]}_{trial_no}.txt"
    gyro_path = f"{root}/{subject}/{category}/{activity}/{activity}_gyro_{subject[3:]}_{trial_no}.txt"

    acc  = load_mobifall_file(acc_path)
    gyro = load_mobifall_file(gyro_path)

    t0 = max(acc["t_ns"].iloc[0], gyro["t_ns"].iloc[0])
    t1 = min(acc["t_ns"].iloc[-1], gyro["t_ns"].iloc[-1])

    step_ns = int(1e9 / target_hz)
    grid = np.arange(t0, t1, step_ns)

    def interp_xyz(df, grid):
        return np.stack([
            np.interp(grid, df["t_ns"], df["x"]),
            np.interp(grid, df["t_ns"], df["y"]),
            np.interp(grid, df["t_ns"], df["z"]),
        ], axis=1)

    acc_rs  = interp_xyz(acc,  grid)
    gyro_rs = interp_xyz(gyro, grid)
    return np.concatenate([acc_rs, gyro_rs], axis=1), grid

X, t = load_trial(ROOT, "sub2", "FALLS", "FOL", 1)
print("Shape:", X.shape, "→ (samples, 6 channels: ax, ay, az, gx, gy, gz)")
print(f"Duration: {(t[-1] - t[0]) / 1e9:.2f}s at ~{len(t) / ((t[-1] - t[0]) / 1e9):.0f} Hz")
print("\nFirst 3 rows:")
print(pd.DataFrame(X[:3], columns=["ax","ay","az","gx","gy","gz"]).round(3))
print("\nStats per channel:")
print(pd.DataFrame(X, columns=["ax","ay","az","gx","gy","gz"]).describe().round(2))
```

```python
def window_trial(X, win_size, stride, max_windows=30):
    T = X.shape[0]
    if T < win_size:
        return np.empty((0, win_size, X.shape[1]), dtype=X.dtype)
    starts = list(range(0, T - win_size + 1, stride))
    if len(starts) > max_windows:
        idx = np.linspace(0, len(starts) - 1, max_windows).astype(int)
        starts = [starts[i] for i in idx]
    return np.stack([X[s:s+win_size] for s in starts], axis=0)

def build_windowed_dataset(all_trials, win_seconds, fs=100, overlap=0.5, max_windows=30):
    win_size = int(win_seconds * fs)
    stride = int(win_size * (1 - overlap))

    X_list, y_activity, y_category, y_subject, y_trial = [], [], [], [], []
    for tr in all_trials:
        windows = window_trial(tr["X"], win_size, stride, max_windows=max_windows)
        n = len(windows)
        if n == 0: continue
        X_list.append(windows)
        y_activity.extend([tr["activity"]] * n)
        y_category.extend([tr["category"]] * n)
        y_subject.extend([tr["subject"]] * n)
        y_trial.extend([(tr["subject"], tr["activity"], tr["trial"])] * n)

    X = np.concatenate(X_list, axis=0)
    return {
        "X": X,
        "y_activity": np.array(y_activity),
        "y_category": np.array(y_category),
        "y_subject": np.array(y_subject),
        "y_trial": y_trial,
        "win_seconds": win_seconds,
    }

for ws in [1, 2, 3]:
    ds = build_windowed_dataset(all_trials, win_seconds=ws)
    print(f"\nWindow = {ws}s  → X shape: {ds['X'].shape}")
    from collections import Counter
    per_act = Counter(ds["y_activity"])
    print(f"  Falls: {sum(v for k,v in per_act.items() if k in ['FOL','FKL','BSC','SDL'])}  "
          f"ADL: {sum(v for k,v in per_act.items() if k not in ['FOL','FKL','BSC','SDL'])}")
    print(f"  Per activity: {dict(sorted(per_act.items()))}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

def load_npz(path, task="binary"):
    d = np.load(path, allow_pickle=True)
    X = d["X"]
    y_subject = d["y_subject"]

    if task == "binary":
        y = d["y_binary"]
    elif task == "fall4":
        mask = d["fall_mask"]
        X, y, y_subject = X[mask], d["y_fall4"][mask], y_subject[mask]
    elif task == "multi13":
        y = d["y_multi13"]
    else:
        raise ValueError(task)

    return X, y, y_subject

def make_splits(X, y, y_subject, seed=42):
    idx = np.arange(len(X))
    trainval_idx, test_idx = train_test_split(
        idx, test_size=0.2, stratify=y, random_state=seed
    )
    train_idx, val_idx = train_test_split(
        trainval_idx, test_size=0.2,
        stratify=y[trainval_idx], random_state=seed
    )
    return train_idx, val_idx, test_idx

def fit_scaler(X_train):
    mean = X_train.reshape(-1, X_train.shape[-1]).mean(axis=0)
    std  = X_train.reshape(-1, X_train.shape[-1]).std(axis=0) + 1e-8
    return mean, std

def apply_scaler(X, mean, std):
    return (X - mean) / std

X, y, y_subj = load_npz("/content/mobifall_processed/mobifall_w2s.npz", task="binary")
tr, va, te = make_splits(X, y, y_subj)
mean, std = fit_scaler(X[tr])
X_tr = apply_scaler(X[tr], mean, std)
X_va = apply_scaler(X[va], mean, std)
X_te = apply_scaler(X[te], mean, std)

print(f"X: {X.shape}  y dist: {np.bincount(y)}")
print(f"Train: {X_tr.shape}, y dist: {np.bincount(y[tr])}")
print(f"Val:   {X_va.shape}, y dist: {np.bincount(y[va])}")
print(f"Test:  {X_te.shape}, y dist: {np.bincount(y[te])}")
print(f"Mean per channel: {mean.round(3)}")
print(f"Std per channel:  {std.round(3)}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

def build_ann(input_shape, n_classes):
    m = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(n_classes if n_classes > 2 else 1,
                     activation="softmax" if n_classes > 2 else "sigmoid"),
    ], name="ANN")
    return m

def build_cnn(input_shape, n_classes):
    m = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv1D(64, 5, padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPool1D(2),
        layers.Conv1D(128, 5, padding="same", activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPool1D(2),
        layers.Conv1D(128, 3, padding="same", activation="relu"),
        layers.GlobalAveragePooling1D(),
        layers.Dropout(0.3),
        layers.Dense(64, activation="relu"),
        layers.Dense(n_classes if n_classes > 2 else 1,
                     activation="softmax" if n_classes > 2 else "sigmoid"),
    ], name="CNN")
    return m

def build_lstm(input_shape, n_classes):
    m = models.Sequential([
        layers.Input(shape=input_shape),
        layers.LSTM(64, return_sequences=True),
        layers.Dropout(0.3),
        layers.LSTM(64),
        layers.Dropout(0.3),
        layers.Dense(32, activation="relu"),
        layers.Dense(n_classes if n_classes > 2 else 1,
                     activation="softmax" if n_classes > 2 else "sigmoid"),
    ], name="LSTM")
    return m

def compile_model(m, n_classes):
    if n_classes > 2:
        loss = "sparse_categorical_crossentropy"
    else:
        loss = "binary_crossentropy"
    m.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss=loss, metrics=["accuracy"])
    return m

for builder in [build_ann, build_cnn, build_lstm]:
    m = compile_model(builder((200, 6), 2), 2)
    print(f"\n=== {m.name} ===")
    m.summary(line_length=80)
```

```python
import time
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

def train_one(build_fn, X_tr, y_tr, X_va, y_va, X_te, y_te, epochs=50, batch=64, seed=42):
    tf.random.set_seed(seed); np.random.seed(seed)

    n_classes = int(max(y_tr.max() + 1, 2))
    model = compile_model(build_fn(X_tr.shape[1:], n_classes), n_classes)

    cbs = [
        callbacks.EarlyStopping(patience=8, restore_best_weights=True, monitor="val_loss"),
        callbacks.ReduceLROnPlateau(patience=4, factor=0.5, monitor="val_loss", min_lr=1e-5),
    ]
    t0 = time.time()
    hist = model.fit(X_tr, y_tr, validation_data=(X_va, y_va),
                     epochs=epochs, batch_size=batch, verbose=0, callbacks=cbs)
    train_time = time.time() - t0

    probs = model.predict(X_te, verbose=0)
    if n_classes == 2:
        y_pred = (probs.ravel() > 0.5).astype(int)
    else:
        y_pred = probs.argmax(axis=1)

    acc = accuracy_score(y_te, y_pred)
    f1  = f1_score(y_te, y_pred, average="macro")
    cm  = confusion_matrix(y_te, y_pred)

    return {
        "model": build_fn.__name__.replace("build_", "").upper(),
        "acc": acc, "f1_macro": f1, "cm": cm,
        "epochs_ran": len(hist.history["loss"]),
        "train_time_s": train_time,
        "history": hist.history,
    }

results = []
for ws in [1, 2, 3]:
    X, y, y_subj = load_npz(f"/content/mobifall_processed/mobifall_w{ws}s.npz", task="binary")
    tr, va, te = make_splits(X, y, y_subj)
    mean, std = fit_scaler(X[tr])
    X_tr, X_va, X_te = apply_scaler(X[tr], mean, std), apply_scaler(X[va], mean, std), apply_scaler(X[te], mean, std)
    y_tr, y_va, y_te = y[tr], y[va], y[te]

    for build_fn in [build_ann, build_cnn, build_lstm]:
        print(f"→ Training {build_fn.__name__} on w={ws}s ...", end=" ", flush=True)
        r = train_one(build_fn, X_tr, y_tr, X_va, y_va, X_te, y_te)
        r["window_s"] = ws
        results.append(r)
        print(f"acc={r['acc']:.4f}  f1={r['f1_macro']:.4f}  "
              f"epochs={r['epochs_ran']}  time={r['train_time_s']:.1f}s")

df = pd.DataFrame([{k: v for k, v in r.items() if k not in ["cm", "history"]}
                   for r in results])
df = df[["window_s", "model", "acc", "f1_macro", "epochs_ran", "train_time_s"]]
print("\n=== Binary classification results ===")
print(df.to_string(index=False))

import pickle
with open("/content/mobifall_processed/task3_binary_results.pkl", "wb") as f:
    pickle.dump(results, f)
print("\nResults saved to task3_binary_results.pkl")
```
